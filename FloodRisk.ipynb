# Data Analysis

The data is provided from two different categories where one is FEMA and the other is First Street. The interactive dashboard is designed to display flood danger in the past, present, and future, paying close attention to current and anticipated environmental changes that could increase flood risk. Given that it is the first data of its sort to be made available for free across the country, the public's access to this information is advantageous to both property owners and the general public.

At a more detailed level, the findings highlight how unevenly different parts of the nation will be affected by shifting environmental circumstances. They also show that in order to adequately comprehend flood risk, additional localized data at the property level are required. The results of the Flood Factor at the property-level seem substantially different from how risk is viewed at the city, county, or state level. The Flood Factor, which ranges from 1 to 10, is a measure of a property's actual flood danger. Higher Flood Factors increase the likelihood that a property will flood, experience heavy floods, or both. The possible depth of a flood as well as the chance of flooding affect a property's Flood Factor.

The number of properties at any risk* flooding is determined using the First Street Foundation Flood Model. For instance, the research indicates that 439,300 properties in Alabama are at risk over the next 30 years when considering this higher degree of risk. 94,000 of these properties were classified as virtually certainly at risk, with a 99% likelihood that flooding will occur at least once over the following 30 years.
